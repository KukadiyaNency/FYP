```python
# Standardise date format
RW.iloc[:, 0] = pd.to_datetime(RW.iloc[:, 0], dayfirst=True, errors='coerce').dt.strftime('%d/%m/%Y')
```

```python
# Replace '*' with NaN
RW.replace('*', np.nan, inplace=True)
```

```python
# Replace spaces with underscores in column names
RW.columns = RW.columns.str.replace(" ", "_")
```

```python
# Convert percentage columns to float
cols_to_convert = RW.columns[1:]  # Exclude 'Date' column
RW[cols_to_convert] = RW[cols_to_convert].replace('%', '', regex=True).astype(float)
```

```python
# Get the percentage of missing values in each column.
missing_pct = round(RW.isnull().sum()/len(RW)*100,1)
print(missing_pct)
```

```python
RW.ffill(inplace=True)  # First, try forward fill
RW.bfill(inplace=True)  # Then, backward fill as a backup
RW.interpolate(method='linear', inplace=True)  # Finally, interpolate remaining gaps
```

```python
RW
```

```python
# Save cleaned file
cleaned_file_path = "/content/RW_cleaned.csv"
RW.to_csv(cleaned_file_path, index=False)
```

```python
RW = pd.read_csv("/content/drive/MyDrive/Cleaned_FYP/RW_cleaned.csv")
```

```python
RW.describe()
```
